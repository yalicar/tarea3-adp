```python
nombre_archivo= "Advertising.csv"
class regresion():
    
     #Variable de clase nombre del archivo a trabjar
    x=[]                               #Variable de clase lista vacia
    y=[]                               #Variable de clase lista vacia
    
    def __init__(self,xx,yy):
        self.x=xx
        self.y=yy
        self.beta0 =(sum([x*x for x,x in zip(self.x,self.x)])*sum(self.y)-sum(self.x)*sum([x*y for x,y in zip(self.x,self.y)]))/(len(self.x)*sum([x*x for x,x in zip(self.x,self.x)])-sum(self.x)*sum(self.x))
        self.beta1=(len(self.x)*sum([x*y for x,y in zip(self.x,self.y)])-(sum(self.x)*sum(self.y)))/(len(self.x)*sum([x*x for x,x in zip(self.x,self.x)])-sum(self.x)*sum(self.x))
        
    
    #Metodo para predecir un valor
    def predict(self,val):
        prediccion = self.beta0+self.beta1*val
        return prediccion
    
    #Medoto par apredecir N Valores
    def predictN(self,vals):
        predn=[] #definiendo una lista vacia 
        for i in range(len(vals)):
            predn.append(self.beta0+self.beta1*vals[i]) # agregando cada valor de la predicicon caluclada a la lista
        return predn
    
    #Metodo para calcular r2
    def r2(self):
        Ycal=[] #lista vacia para almacenar el valor de Ycalculada
        err=[]  #Lista vacia para almacenar el error
        err2=[]
        sYcal=0  # Suma de todos los valores de Y
        serr2=0  # Promedio del error cuadrado
        den=0 #Denominador de la formula
        for i in range(len(self.x)):
            Ycal.append(self.beta0 + self.beta1*self.x[i])
            err.append(self.y[i]-Ycal[i])
            err2.append(err[i]*err[i])
            sYcal=self.y[i]+sYcal
            sYcalmed=float(sYcal/len(self.x))
            serr2=err2[i]+serr2
        for i in range(len(self.y)):
            den=((self.y[i]-sYcalmed)**2)+den
        r2 = (1-(serr2/den))
        return (r2)
        
    # Metodo para calcular la raiz cuadrada de r2.
    def r(self):
        numero=self.r2()
        r=numero
        i=0
        while i!=r:
            i=r
            r=(numero/r+r)/2
        return r
    # Metodo que devuelve los parametros de la regresion
    def getAllRegParams(self):
        dic={"b0":self.beta0,"b1":self.beta1,"r2":self.r2(),"r":self.r()}
        return dic # falta verificar que lo que devuelve es un diccionario
    
    #Decorador para leer archivo csv.
    @classmethod 
    def from_file(cls,path):
        xx=[] # Creando lista X vacia
        yy=[] # Creando lista Y vacia
        z=0
        sx2=0
        sxi2=0
       #newline=''
        with open(path,'r') as arch:
            next(arch)
            yy, xx = zip(*((float(n) for n in line.rstrip().rsplit(",", maxsplit=2)[-2:]) for line in arch))
            return cls(xx,yy)  
        
wer=[5,6,90,6]       
pedro = regresion.from_file(nombre_archivo)
print(pedro.predict(250))
print(pedro.getAllRegParams())
print(pedro.predictN(wer))



   
```
